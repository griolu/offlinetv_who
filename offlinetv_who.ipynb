```
from fastai.vision.all import *
from fastai.vision.widgets import *
```

# OfflineTV Who?
New to watching OfflineTV or don't care who's who? Use this model to check what member is in the photo. The model is overfitted on images of Scarra, Pokimane, LilyPichu, DisguisedToast, Yvonnie, and Michael Reeves... poor Brodin Plett :(. Upload an image containing one person and see who they are! 

```
# Import the .pkl file
path = Path()
learn_inf = load_learner(path/'export.pkl')
```

```
# Create all the widgets
upload_btn = widgets.FileUpload()
out_img = widgets.Output()
lbl_pred = widgets.Label()
run_btn = widgets.Button(description = 'Predict')
```

```
# Button click function
def run_btn_clicked(change):
  if upload_btn.data == None:
    return
  img = PILImage.create(upload_btn.data[-1])
  out_img.clear_output()
  with out_img: display(img.to_thumb(128, 128))
  pred, pred_idx, probs = learn_inf.predict(img)
  if pred_idx == 0:           # Disguised Toast
    name = "Disguised Toast"
  elif pred_idx == 1:         # LilyPichu
    name = "LilyPichu"
  elif pred_idx == 2:         # Michael Reeves
    name = "Michael Reeves"
  elif pred_idx == 3:         # Pokimane
    name = "Pokimane"
  elif pred_idx == 4:         # Scarra
    name = "Scarra"
  else:                       # Yvonnie
    name = "Yvonnie"
  lbl_pred.value = f'I think it\'s {name}, with a {probs[pred_idx]:.04f} probability.'

run_btn.on_click(run_btn_clicked)
```

```
# VBox for the widgets
VBox([widgets.Label('Upload the image:'),
      upload_btn, run_btn, out_img, lbl_pred])
```
